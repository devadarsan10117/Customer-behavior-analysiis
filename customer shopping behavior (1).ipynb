```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(
    lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum() 
```

```python
df.columns = df.columns.str.lower()
df.columsn = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#create a column age_group
labels = ['Young Adult','Adult','Middle-age','Senior']
df['age_group'] = pd.qcut(df['age'], q=4,labels=labels)
```

```python
df[['age','age_group']].head(10) 
```

```python
frequency_mapping = {
    'Bi-Weekly': 14,
    'Annually': 365,
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Every 3 Months': 90,
}

```

```python
df['purchase_frequency_days'] = df['frequency of purchases'].map(frequency_mapping)

```

```python
# Display 10 rows of the original and mapped columns
df[['frequency of purchases', 'purchase_frequency_days']].head(10)
```
